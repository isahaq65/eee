```python
import numpy as np
import matplotlib.pyplot as plt
```

 $$Sinusoidal Signal$$
 Sinusoidal Signals are periodic functions which are based on the sine or cosine function from trigonometry.The general form of a Sinusoidal Signal


> $x(t)=A sin(ωt +ϕ)$     
>$x(t)=A cos(ωt +ϕ)$

ωt +ϕ or 2πft +ϕ is angle (in radians) of the sine function
 Since the angle depends on time, it makes x(t) a signal
ω is the radian frequency of the sinusoidal signal<br>
 f is called the cyclical frequency of the sinusoidal signal<br>
 ϕ is the phase shift or phase angle<br>
A is the amplitude of the signal<br>





**$$x(t)=Asin(2\pi f t+ϕ)$$**

***Plotting  x(t) sinusoidal signal in continuous, stem, scatter plot***

```python
t=np.linspace(0, 4, 160)
```

```python
def generate_sinusoidal_signal(A, f, initial_phase=0):
  y = A * np.sin(2.0 * np.pi * f * t + initial_phase)
  return y
```

```python
x = generate_sinusoidal_signal(1,1)

fig, position = plt.subplots(nrows=2, ncols=2, figsize=(18, 15))
fig.suptitle(' plots of a sinusoidal signal', fontsize=15)

position[0, 0].plot(t,x)
position[0, 0].set_title('Continuous plot', fontsize = 15)

position[0, 1].stem(t,x)
position[0, 1].set_title('Stem plot', fontsize = 15)

position[1,0].scatter(t,x)
position[1,0].set_title('Scatter plot', fontsize = 15)
position[-1,-1].axis('off')



plt.show()
```

**Plotting three sinusoidal signal with different properties**

```python
x1 = generate_sinusoidal_signal(2,2)
```

```python
x2 = generate_sinusoidal_signal(3, 5)
```

```python
x3 =generate_sinusoidal_signal(3, 0.5)
```

```python
fig, position = plt.subplots(nrows=2, ncols=2, figsize=(18, 15))
fig.suptitle('Three different sinusoid signals', fontsize=15)

position[0, 0].plot(t,x1)
position[0, 0].set_title('x1(t)', fontsize = 15)

position[0, 1].plot(t,x2, 'r')
position[0, 1].set_title('x2(t)', fontsize = 15)

position[1,0].plot(t,x3,'g')
position[1,0].set_title('x3(t)', fontsize = 15)
position[-1,-1].axis('off')



plt.show()
```

**Composite Signal**<br>
A composite signal can be periodic or non-periodic. A periodic composite signal can be decomposed into a series of simple sine waves with discrete frequencies, frequencies that have integer values (1, 2, 3, and so on).




**Composite signal of three given signals**

```python
plt.plot( t, x1+x2+x3)
plt.title('Composite Signal')
plt.show()
```
